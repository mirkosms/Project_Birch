```python
from sklearn.datasets import load_digits
from sklearn.metrics import homogeneity_score, completeness_score
import matplotlib.pyplot as plt
import numpy as np

# Zaaduj zbir danych
data = load_digits()
X = data.data
y = data.target  # Etykiety (0-9)

# Zaaduj algorytm (Twoja implementacja)
from birch import Birch

# Parametry (dostosowane na podstawie testw)
threshold = 12
branching_factor = 40

# Uruchom algorytm
birch = Birch(threshold=threshold, branching_factor=branching_factor)
birch.fit(X)
labels = birch.predict(X)

# Ocena wynikw
print("Homogeneity Score:", homogeneity_score(y, labels))
print("Completeness Score:", completeness_score(y, labels))

# Opcjonalna wizualizacja
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)

plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=labels, cmap='viridis', s=10)
plt.title("Clusters by Birch")
plt.savefig(f'clusters_threshold_{threshold}_branching_{branching_factor}.png', dpi=300, bbox_inches='tight')
plt.show()
```
